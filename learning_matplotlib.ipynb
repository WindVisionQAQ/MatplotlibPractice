### Load Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize figure

plt.figure(figsize=(5,3),dpi=100)

# Line 1
# Use shorthand notation
# fst = '[color][marker][linestyle]'
plt.plot(x,y,'r.-',label='2x')
plt.title('Our First Graph!',fontdict={'fontname':'Comic Sans MS','fontsize':20})



# Line 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,label='x^2')
plt.plot(x2[5:],x2[5:]**2,color='blue',linestyle='--')

# Add label to both axis
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()

# Save image
plt.savefig('mygraph.png',dpi=300)
```

### Bar plot

```python
label = ['A','B','C']
value = [1,4,2]

bar = plt.bar(label,value)
bar[0].set_hatch('/')
bar[1].set_hatch('O')
bar[2].set_hatch('*')
plt.show()
```

### Gas Price line plot

```python
gas = pd.read_csv("gas_prices.csv")


plt.figure(figsize=(8,5))
# add title
plt.title('Gas price over time')

plt.plot(gas.Year,gas.USA,'r.-',label='USA')
plt.plot(gas.Year,gas.Canada,'g.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'b.-',label='South Korea')
plt.plot(gas.Year,gas.Australia,'y.-',label='Australia')


# add legend
plt.legend()

# add description to axis
plt.xlabel('Year')
plt.ylabel('USD')

# adjust x axis
plt.xticks(gas.Year[::3])


plt.show()

plt.savefig("gas_prices.png",dpi=300)
```

### Load fifa data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

### Histgram 

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall,bins=bins)
plt.xticks(bins)

plt.xlabel('Player Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of players on skill level')

plt.show()
plt.savefig("player_hist.png",dpi=300)
```

### Pie chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
```

```python
plt.pie([left,right],labels=['left','right'],autopct='%.2f %%')
plt.title('Foot Preference of Fifa players')
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]
```

```python
light = fifa.loc[fifa.Weight<125].count()[0]
medium_light = fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]

Weights = [light,medium_light,medium,medium_heavy,heavy]
labels = ['<125','125-150','150-175','175-200','>200']
explodes = [0.4,0,0,0,0.4]

plt.style.use('ggplot')
plt.figure(figsize=(10,10))
plt.pie(Weights,labels=labels,explode=explodes)
plt.title('Distribution of weight of fifa players')
plt.show()
```

```python
fcb = fifa.loc[fifa.Club == 'FC Barcelona','Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid','Overall']
mc = fifa.loc[fifa.Club == 'Manchester City','Overall']
```

```python
teams = [fcb,madrid,mc]
labels = ['FC Barcelona','Real Madrid','Manchester City']

plt.style.use('default')
plt.boxplot(teams,labels=labels)
plt.title('Skill level distribution of teams in FIFA')
plt.xlabel('Team')
plt.ylabel('Skill Level')

plt.show()
```
